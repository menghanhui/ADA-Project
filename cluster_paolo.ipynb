```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mlxtend
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import re
```

```python
italian_tweets_text = pd.read_csv('italian_tweet_text_notusa.csv',header=None)
```

```python
stopwords = pd.read_json('stopwords-it.json')
```

```python
italian_tweets_text.head()
```

```python
stopwords = stopwords[0].tolist()
```

```python
#stopwords
```

```python
italian_tweets_text.shape[0]
```

```python
documents=italian_tweets_text[1].tolist()
documents=[re.sub(r'(\s)http\w+', r' ', m.lower()) for m in documents]
documents
```

```python
%%capture
vectorizer = TfidfVectorizer(stop_words=stopwords)
X = vectorizer.fit_transform(documents)
```

```python
true_k = 8
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=10000, n_init=50)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
common_words = model.cluster_centers_.argsort()[:,-1:-21:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(terms[word] for word in centroid))
```

```python
variety_df=pd.DataFrame()
variety_df['cluster'] = model.labels_
```
