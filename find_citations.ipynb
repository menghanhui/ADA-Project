```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from zipfile import ZipFile
```

```python
data_folder = './data/'
zip_file = ZipFile(data_folder+'russian-troll-tweets.zip')
fs = {text_file.filename: pd.read_csv(zip_file.open(text_file.filename))
       for text_file in zip_file.infolist()
       if text_file.filename.endswith('.csv')}
```

```python
data=pd.DataFrame()
for key,val in fs.items():
    data=pd.concat([data,val])
```

```python
italian_tweets = data[data.language == 'Italian']
italian_tweets['publish_date'] =  pd.to_datetime(italian_tweets['publish_date'], format='%m/%d/%Y %H:%M')
```

```python
retweet_italian_tweets = italian_tweets.loc[italian_tweets['post_type'] == 'RETWEET']
retweet_italian_tweets
```

```python
retweet_italian_content = retweet_italian_tweets.loc[retweet_italian_tweets['content'].str.contains('@')]
retweet_italian_content.content
```

```python
#retweet_italian_content['content']

def returncit(text):
    final = []
    list_words= text.split()
    for word in list_words:
        if word.find('@') > -1:
            list_cit = word.split('@')
            for cit in list_cit:
                if (cit != '') and (len(cit)>2):
                    final.append(cit)
    return final
```

```python
list_cit = retweet_italian_content.content.apply(returncit)
```

```python
retweet_italian_content['citations'] = list_cit
```
