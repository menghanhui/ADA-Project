```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
italian_tweets = pd.read_csv('italian_tweets_old.csv')
```

```python
italian_tweets['publish_date'] =  pd.to_datetime(italian_tweets['publish_date'], format='%m/%d/%Y %H:%M')
```

```python
italian_content = italian_tweets.loc[italian_tweets['content'].str.contains('@')]
#italian_content.content
```

```python
regex = re.compile('[^a-zA-Z_]')
```

```python
#retweet_italian_content['content']

def returncit(text):
    final = []
    list_words= text.split()
    for word in list_words:
        if word.find('@') > -1:
            list_cit = word.split('@')
            for cit in list_cit:
                if cit != '':
                    cit = regex.sub('', cit)
                    if cit != '':
                        final.append(cit)
                                       
    return final
```

```python
list_cit = italian_content.content.apply(returncit)
```

```python
italian_content['citations'] = list_cit
```
